### Appendix: `get_pretrained_word_embed(self)` method

```python
from pymagnitude import *
glove = Magnitude('GoogleNews-vectors-negative300.magnitude')
```

```python
from pymagnitude import *

def get_pretrained_word_embed(vocab, magnitude, verbose=1):
    """
    loads word embeddings from magnitude into pytorch
    """
    if type(magnitude)==str:
        magnitude = Magnitude(magnitude)
    elif type(magnitude)==pymagnitude.Magnitude:
        pass
    else:
        print("magnitude type not recognized")
        return None, None
    
    pretrained_word_embed = [torch.from_numpy(magnitude.query(word)).type(torch.cuda.FloatTensor).view(1,-1)
                             for word in vocab]
    pretrained_word_embed = torch.cat(pretrained_word_embed, axis=0)
    
    unknown_words = [word in magnitude for word in vocab]
    n_unknown_words = sum(unknown_words)
    if verbose: print(f"Pretrained embeddings loaded: {n_unknown_words} / {len(vocab)}")
    
    return pretrained_word_embed, unknown_words

pretrained_word_embed, unknown_words = get_pretrained_word_embed(vocab, glove)
```
